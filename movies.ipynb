# Comparison of Bollywood and Hollywood Movies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the data in csv format
ratings = pd.read_csv('All_Movies1.csv', engine='python')
```

```python
# filter the movies as bollywood and hollywood
bollywood_movies = ratings[ratings['original_language'] == 'hi']
hollywood_movies = ratings[ratings['original_language'] == 'en']
```

```python
# calculates the mean of each year
bollywood_mean = bollywood_movies.groupby('release_date')['vote_average'].mean()
hollywood_mean = hollywood_movies.groupby('release_date')['vote_average'].mean()
```

```python
# added the data after plotting for readability
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
bollywood_rating = [5.1, 5.4, 5.1, 5.6, 5.65, 5.3, 6.4, 5.9, 5.5, 6.1, 5.85]
hollywood_rating = [6.28, 6.28, 6.15, 6.38, 6.35, 6.4, 6.53, 6.53, 6.45, 6.55, 6.53]

plt.figure(figsize=(8, 6))
plt.plot(year, bollywood_mean, label='Bollywood', marker='o', linestyle='-', color='blue')
plt.plot(year, hollywood_mean, label='Hollywood', marker='o', linestyle='-', color='red')

plt.grid(True)
plt.title('Mean Rating for Movies')
plt.xlabel('Release Year')
plt.ylabel('Mean Ratings')

for i in range(len(year)):
    plt.text(year[i], bollywood_rating[i], str(bollywood_rating[i]), ha='center', va='bottom')
    plt.text(year[i], hollywood_rating[i], str(hollywood_rating[i]), ha='center', va='bottom')

plt.legend()   
plt.show()
```
